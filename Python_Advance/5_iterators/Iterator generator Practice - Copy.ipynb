#### Iterator Basics

In directory there is a file **pricing.txt**. 

Read this file and create a list of prices.

get prices[:2].

**Expected Output:**

[{'ticker': 'MSFT', 'country': 'BRA', 'price': 381.26, 'date': '2023-10-27'},

 {'ticker': 'CSCO', 'country': 'USA', 'price': 187.64, 'date': '2023-02-12'}]

```python
#write your code here
prices= []
with open("pricing.txt","r") as file:
    for line in file:
        ticker, country, price, date= line.split("|")
        prices.append({
            "ticker":ticker,
            "country":country,
            "price":price,
            "date":date.strip()})
prices[:2]
```

### Read the file (pricing.txt)
By using conventional approach

```python
#write your code here
with open("pricing.txt","r") as file:
    content=file.read()
print(content)
```

### Read the file
By using line by line approach

```python
#write your code here
with open("pricing.txt", "r") as file:
    for line in file:
        print(line, end="")
```

### Read the file
By using list approach

```python
#write your code here
with open("pricing.txt","r") as file:
    content=file.readlines()
print(content)
```

### Read the file by using
generator approach

```python
#write your code here
def get_price():
    with open("pricing.txt","r") as file:
        content=file.read()
        yield content
for price in get_price():
    print(price)
    
```
