#### Iterator Basics

In directory there is a file **pricing.txt**. 

Read this file and create a list of prices.

get prices[:2].

**Expected Output:**

[{'ticker': 'MSFT', 'country': 'BRA', 'price': 381.26, 'date': '2023-10-27'},

 {'ticker': 'CSCO', 'country': 'USA', 'price': 187.64, 'date': '2023-02-12'}]

```python
#write your code here
prices=[]
with open("pricing.txt","r") as file:
    for line in file:
        ticker,country,price,date=line.split("|")
        prices.append({
            "ticker":ticker,
            "country":country,
            "price":price,
            "date":date.strip()})
print(prices)
```

```python
prices[:2]
```

### Read the file
By using line by line approach

```python
#write your code
with open("pricing.txt","r") as file:
    for line in file:
        print(line,end="")
```

### Read the file
By using list approach

```python
#write your code here
with open("pricing.txt","r") as file:
    data=file.readlines()
data
```

### Read the file by using
generator approach

```python
#write your code here
def get_data():
    with open("pricing.txt","r") as file:
        data=file.read()
        yield data
for data in get_data():
    print(data)
```

### Stock files
There are three stock files in the directory.
Read those files together by using traditional approach

```python
#write your code here
stocks=[]
for i in range(3):
    with open(f"stock_{i+1}.txt","r") as file:
        data=file.read()
        stocks.append(data)
print(stocks)
```

### DO the above by generator approach
create a list of all those stocks

```python
#write your code here
def get_company_data():
    for i in range(3):
        with open(f"stock_{i+1}.txt","r") as file:
            data=file.read()
            yield data
for company in get_company_data():
    print(company)
```

```python
import json
company_data= [json.loads(company) for company in get_company_data()]
company_data[1]["price"]
```

#### Fibonacci Sequence Generator

Create a Fionacci series by using generator approach

```python
def get_fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a,b=b,a+b

fib_gen=get_fibonacci()
for i in range(10):
    print(next(fib_gen))
```
