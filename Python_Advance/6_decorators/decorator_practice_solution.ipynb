#### Timing functions without decorators

Create a function to get square of given numbers.

Also check the time duration (End Time-Start time) required for the processing.

```python
import time
def get_square(*args):
    start=time.time()
    result=[]
    for n in args:
        result.append(n*n)
    end=time.time()
    print(f"The time duration for processing is {end-start} seconds")
    return result
    
```

```python
get_square(2,4,5)
```

#### Timing functions with decorator

Write a same thing above with decorator function

```python
def timer(func):
    def wrapper(*args,**kwargs):
        start=time.time()
        result=func(*args, **kwargs)
        end=time.time()
        print(f"Function {func. __name__} took {end-start} seconds")
        return result
    return wrapper
```

```python
@timer
def get_square(*args):
    result=[]
    for n in args:
        result.append(n*n)
    return result
```

```python
get_square(2,4,5)
```
